## Dropping Columns That Are Not Relevant To The Analysis

### I will be Dropping the Columns 'Daily Summary', and 'Loud Cover'. 
### Since in my initial Analysis, I feel that the "Summary" Column is providing enough information for the Weather of the Day, so The Daily Summary Column feels Redundant, Hence I will be dropping it. 
### As For The "Loud Cover" column, It has all values as 0, and so it serves no meaningful purpose to the Data, so I will be dropping it as well.

```python
df.drop(['Daily Summary','Loud Cover'],axis=1,inplace=True)
df
```

## Analyzing The Data To Check For Missing Values

```python
df.columns[df.isnull().any()] # This Shows That 'Precip Type' Column has Missing Values
```

```python
df.info() # This Also Shows That 'Precip Type' Column has Missing Values (95936 instead of 96453)
```

```python
df[df.isnull().any(axis=1)] # This Shows All Entries with Missing Values
```

### Filling The Missing Data with "Not Defined" Keyword, in-order to use this 'Precip Type' Column for Later Analysis.

```python
df.fillna("Not Defined", inplace=True)
df
```

```python
df[df.isnull().any(axis=1)] # This Shows that there are no longer missing values in any column of our DataFrame.
```

```python
df.info() # This Also Shows that there are no longer missing values in any column of our DataFrame.
```

```python
df.shape # This Shows the Current Status or "Shape" of our DataFame.
```

### Resampling the Data With Respect to "Months".

```python
df[['Formatted Date']] # Initial State of Dates in DataFrame
```

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
df
df_2=df.set_index('Formatted Date')
df_2 = df_2.resample('M').mean()
df_2 

# This is the breakdown of Data With Respect to Month. It Accounts for 11 Years and 1 Month of Data. [Extra Month of 2005]
```

```python
df_2.info()
```

```python
df_2 # This is the Data we have left to apply Visualization, Resampled into Months.
```

## Time for Analysis Phase, With Use of Visualization!

### I will be using Seaborn Library for most of my Visualization, Mostly Because It has more customization as compared to Matplotlib and personally it feels easier to implement various types of graphs

### Simple Comparision of Apparent Temperature vs. Humidity

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
g = sns.relplot(x='Apparent Temperature (C)', y='Humidity', data=df, kind="scatter", color="purple")
g.fig.suptitle("Apparent Temperature (C) vs. Humidity",y=1.1)
plt.show()

from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

### Comparision of Humidity vs. Wind Speed (km/h) 

### According to Scientific Research, The Higher wind speed causes minimum evaporation of water, and low humidity, or vice versa. We can see the trend being proven in this plot below.

```python
plt.figure(figsize=(12,8))
g = sns.relplot(x='Wind Speed (km/h)',y='Humidity', data=df, kind="line", color="cyan", ci=None)
g.fig.suptitle("Humidity vs. Wind Speed (km/h)",y=1.1)
plt.show()
```

### Comparision of Apparent Temperature vs. Humidity with respect to "Summary" of the weather on a given day using Pair Plots.

```python
g = sns.pairplot(df, vars=['Humidity', 'Apparent Temperature (C)'], hue='Summary')
g.fig.suptitle("Apparent Temperature (C) vs. Humidity w.r.t Summary",y=1.2)
plt.show()
```

### Comparision of Apparent Temperature vs. Humidity with respect to "Summary" of the weather on a given day using Relational Plots with Columns separated as "Summary".

```python
plt.figure(figsize=(16,6))
g = sns.relplot(x='Apparent Temperature (C)',y='Humidity',color='purple',hue="Summary",data=df, col="Summary", col_wrap=3)
g.fig.suptitle("Apparent Temperature (C) vs. Humidity w.r.t Summary",y=1.005)
plt.show()
```
